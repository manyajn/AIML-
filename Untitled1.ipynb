

# **Preprocessing Data**



*   Resizeing the image
*   Converting the image into an array


*   Converting image to RGB fromat
*   Storing images in a list


*   Encoding labels to arrays according to there categories
*   Converting list to numpy arrays, to make them ready for input by the


```python
image_list = []
label_list = []

folder_path = "drive/MyDrive/data"

for j in os.listdir("drive/MyDrive/data"):
  image_path=os.path.join(folder_path, j)
  image_files = [f for f in os.listdir(image_path)]
  for image_file in image_files:
      img = Image.open(os.path.join(image_path, image_file))
      img = img.convert("RGB").resize((256,256))
      img_array = np.array(img)
      image_list.append(img_array/255)
      label_list.append(j)
```

```python
label = []
for i in range(0,len(label_list)):
  if(label_list[i] == "6056"):
    label.append([1,0,0])
  if(label_list[i] == "6060"):
    label.append([0,1,0])
  if(label_list[i] == "6062"):
    label.append([0,0,1])
```

```python
fig, axes = plt.subplots(5, 4, figsize=(15, 18))
a = np.random.randint(0,216,20)
k=0
for i in range(0,5):
  for j in range(0,4):

    axes[i][j].imshow(image_list[a[k]])
    axes[i][j].set_title(label_list[a[k]])
    k=k+1

plt.show()
```

```python
label = np.asarray(label)
image_list = np.asarray(image_list)
```

```python
image_list.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

# **Using SGD optimizer**


```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(3, activation='sigmoid'))
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer=tf.keras.optimizers.SGD(),
              metrics=keras.metrics.Accuracy())
```

```python
initial_hist = model.fit(x=image_list,y=label,batch_size = 32,epochs=20)
```

```python
plt.plot(np.arange(1, 21), initial_hist.history['loss'], label='Loss', lw=3)
plt.plot(np.arange(1, 21), initial_hist.history['accuracy'], label='Accuracy', lw=3)
plt.title('Evaluation metrics', size=20)
plt.xlabel('Epoch', size=14)
plt.legend()
```

# **Using Adam Optimizer**

```python
model1 = Sequential()
```

```python
model1.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model1.add(MaxPooling2D())
model1.add(Conv2D(32, (3,3), 1, activation='relu'))
model1.add(MaxPooling2D())
model1.add(Conv2D(16, (3,3), 1, activation='relu'))
model1.add(MaxPooling2D())
model1.add(Flatten())
model1.add(Dense(256, activation='relu'))
model1.add(Dense(3, activation='sigmoid'))
```

```python
model1.compile(loss="categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
```

```python
initial_hist1 = model1.fit(x=image_list,y=label,batch_size = 32,epochs=7)
```

```python
plt.plot(np.arange(1, 8), initial_hist1.history['loss'], label='Loss', lw=3)
plt.plot(np.arange(1, 8), initial_hist1.history['accuracy'], label='Accuracy', lw=3)
plt.title('Evaluation metrics', size=20)
plt.xlabel('Epoch', size=14)
plt.legend()
```
